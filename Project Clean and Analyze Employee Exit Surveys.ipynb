### Identify Dissatisfied Employees

```python
tafe_resignations['Contributing Factors. Dissatisfaction'].value_counts(dropna = False)
```

```python
tafe_resignations['Contributing Factors. Job Dissatisfaction'].value_counts(dropna = False)
```

```python
def update(val):
    if pd.isnull(val):
        return np.nan
    elif val == '-':
        return False
    else:
        return True
```

```python
tafe_resignations['Contributing Factors. Dissatisfaction'] =\
tafe_resignations['Contributing Factors. Dissatisfaction'].apply(update)

tafe_resignations['Contributing Factors. Job Dissatisfaction'] =\
tafe_resignations['Contributing Factors. Job Dissatisfaction'].apply(update)
```

```python
tafe_resignations['dissatisfied'] = tafe_resignations\
[['Contributing Factors. Dissatisfaction','Contributing Factors. Job Dissatisfaction']].any(axis=1, skipna = False)
```

```python
dete_resignations['dissatisfied'] = dete_resignations[
    ['interpersonal_conflicts',
       'job_dissatisfaction', 'dissatisfaction_with_the_department',
       'physical_work_environment', 'lack_of_recognition',
       'lack_of_job_security', 'work_location', 'employment_conditions',
       'maternity/family', 'relocation', 'study/travel', 'ill_health',
       'traumatic_incident', 'work_life_balance', 'workload',
       'none_of_the_above']
    ].any(axis=1, skipna=False)
```

```python
dete_resignations_up = dete_resignations.copy()
tafe_resignations_up = tafe_resignations.copy()
```

### Combine the Data

```python
dete_resignations_up['institute'] = 'DETE'
tafe_resignations_up['institute'] = 'TAFE'
```

```python
combined = pd.concat([dete_resignations_up, tafe_resignations_up], ignore_index=True)

combined
```

```python
combined_updated = combined.dropna(thresh=500, axis=1).copy()
```

### Clean the Service Column

```python
combined_updated['institute_service'] = combined_updated['institute_service'].astype('str').str.extract(r'(\d+)').astype('float')

combined_updated
```

```python
def career(x):
    if pd.isnull(x):
        return np.nan
    elif x < 2:
        return 'Fresher'
    elif 2 <= x <= 4:
        return 'Junior'
    elif 5 <= x <= 7:
        return 'Senior'
    else:
        return 'Veteran'
```

```python
combined_updated['service_cat'] = combined_updated['institute_service'].apply(career)

combined_updated
```

```python
combined_updated.drop_duplicates(inplace = True)
```

```python
def perc(val, total_resign):
    return val / total_resign * 100
```

```python
total_resignations = combined_updated['id'].count()
combined_updated.groupby(by=['service_cat'], dropna = False).agg(
    dissatisfied_perc = ('id', lambda x: perc(x.count(), total_resignations))
    )
```
